```python
import pandas as pd
from plotnine import ggplot, aes, geom_line, ggtitle
from IPython.display import display

```

```python
#definir o array que corresponde aos anos da serie historica

def gerar_anos(ano_inicial, count):
    vetor_anos = [0] * count
    for x in range(count):
        vetor_anos[x] = ano_inicial + x
    return vetor_anos
```

```python
def selecionar_anos(vetor_anos,):
    vetor_anos_selecionados = [0] * 15
    aux = 0
    for ano in vetor_anos:
        if ano >= 2010:
            vetor_anos_selecionados[aux] = ano
            aux = aux + 1
    return vetor_anos_selecionados
```

```python
# Função para criar o dataset
def criar_dataset(vetor_anos_selecionados, vetor_valores):
    return pd.DataFrame({"ano": vetor_anos_selecionados, "PIB": vetor_valores})
```

```python
# Função para plotar o gráfico
def plotar_grafico(dataset):
    return ggplot(dataset) + aes(x="ano", y="PIB") + geom_line() + ggtitle("Série histórica do PIB brasileiro dos últimos 15 anos")
```

```python
# Função principal
def main():
    ano_inicial = 2000
    count = 25
    vetor_valores = [7.7, 4.0, 1.9, 3, 0.5, -3.5, -3.3, 1.3, 1.8, 1.1, -4.1, 4.6, 3.1, 3.2, 3.5]

    # Gerar e selecionar anos
    vetor_anos = gerar_anos(ano_inicial, count)
    vetor_anos_selecionados = selecionar_anos(vetor_anos)

    # Garantir que vetor_valores também tenha 15 elementos
    vetor_valores = vetor_valores[:15]

    # Criar dataset
    dataset = criar_dataset(vetor_anos_selecionados, vetor_valores)
    print(dataset)

    # Exibir gráfico corretamente no Google Colab
    grafico = plotar_grafico(dataset)
    display(grafico)  # Exibe o gráfico corretamente

# Executar a função principal
main()
```
